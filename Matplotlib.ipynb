```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
y
```

```python
#FUNCTIONAL PLOT
plt.plot(x,y)
#.show can be interpretted as printing the plot 
```

```python
#.show can be interpretted as printing the plot 
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OBJECT ORIENTED METHOD
fig=plt.figure()
#We first created a blank canvas and gave it a name

axes=fig.add_axes([0.1,0.1,0.8,0.8])
#Then added some axes to that blank canvas [left,bottom,width,height]


axes.plot(x,y)
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([1.0,0.1,0.8,0.8])

axes1.plot(x,y,'r')
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Title')


axes2.plot(y,x,'b')
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)

#axes.plot(x,y)
plt.tight_layout()
#this method does its best to make space between subplots
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)


#axes.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
#as we can see axes is a list of subplots we can either iterate through it or use indexing like normal list
axes
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
#legend is used to clarify what each plot represents
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

ax.legend()
```

```python
#legend is used to clarify what each plot represents
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')


# To change the position of legend
#loc uses a loc code that can be found on matplotlib documentation
#loc code 0 represents best position so most of the time uses it
ax.legend(loc=10)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')

#If no loc code can get the job done a tuple can be used to specify the position
ax.legend(loc=(0.1,0.1))
```

```python
#Plot appearance
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

#color argument can take a name of color or a rgb hexcode
ax.plot(x,y,color='green')
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='#FF8C00')#RGB Hex Code
```

```python
#Line width and Line style

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',linewidth=3,alpha=0.5,linestyle='--')
#Default Line width is 1
#Alpha argument controls the transparency of line
#in place of 'linewidth' 'lw' can also be written
#line style argument controls the style of line
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=3,alpha=0.5,ls='-.')
#more options for ls are '-' , ':' ,  'steps'
```

```python
#Markers
```

```python
#x is an array of 11 point to mark those points on plots marker is used
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple', lw = 1, alpha = 1, ls = '-', marker='o', markersize = 10,
        markerfacecolor='yellow',markeredgecolor='green',markeredgewidth='3')
#there are alot of arguments to customize markers
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'purple', lw =2, ls='--')
#To change the limit of axes
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
